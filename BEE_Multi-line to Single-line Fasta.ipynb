```python
def convert_fasta_to_single_line(input_file):
    with open(input_file, 'r') as infile:
        current_header = None
        first_line = True
        
        for line in infile:
            line = line.strip()
            if line.startswith('>'):
                if not first_line:
                    print() 
                print(line)
                current_header = line
                first_line = False
            else:
                print(line, end='', flush=True)
        
        if not first_line and not line.startswith('>'):
            print()
            
input_filename = 'multiline_input.fasta'
convert_fasta_to_single_line(input_filename)
```
